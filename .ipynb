```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de exibição
%matplotlib inline
sns.set(style="whitegrid")
```

## Carregando os dados

```python
online_retail_df = pd.read_excel("C:/Users/marcu/OneDrive/Área de Trabalho/online-rentail/online+retail/Online Retail.xlsx")
wholesale_customers_df = pd.read_csv("C:/Users/marcu/OneDrive/Área de Trabalho/d-wholesale+customers/wholesale+customers/Wholesale customers data.csv")
mall_customers_df = pd.read_csv("C:/Users/marcu/OneDrive/Área de Trabalho/d-mall_customers/archive (1)/Mall_Customers.csv")

display(mall_customers_df.head(), wholesale_customers_df.head(), online_retail_df.head())
```

## Limpeza dos dados

```python
# Obter informações gerais dos datasets
print("Mall Customers Info:")
mall_customers_df.info()
print("\nWholesale Customers Info:")
wholesale_customers_df.info()
print("\nOnline Retail Info:")
online_retail_df.info()

# Checar valores nulos e remover duplicatas
mall_customers_df = mall_customers_df.dropna()
wholesale_customers_df = wholesale_customers_df.dropna()
online_retail_df = online_retail_df.dropna(subset=['CustomerID']).drop_duplicates()

# Exibir estatísticas descritivas
print("\nMall Customers Description:")
display(mall_customers_df.describe())
print("\nWholesale Customers Description:")
display(wholesale_customers_df.describe())
print("\nOnline Retail Description:")
display(online_retail_df.describe())
```

## Análise Univariada e Visualização de Variáveis Chave

```python
# Análise univariada
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Mall Customers dataset
sns.histplot(mall_customers_df['Age'], kde=True, ax=axes[0, 0]).set(title='Distribuição da Idade')
sns.histplot(mall_customers_df['Annual Income (k$)'], kde=True, ax=axes[0, 1]).set(title='Distribuição da Renda Anual (k$)')
sns.histplot(mall_customers_df['Spending Score (1-100)'], kde=True, ax=axes[0, 2]).set(title='Distribuição da Pontuação de Gasto (1-100)')

# Wholesale Customers dataset
sns.histplot(wholesale_customers_df['Fresh'], kde=True, ax=axes[1, 0]).set(title='Produtos Frescos - Distribuição')
sns.histplot(wholesale_customers_df['Milk'], kde=True, ax=axes[1, 1]).set(title='Produtos de Leite - Distribuição')
sns.histplot(wholesale_customers_df['Grocery'], kde=True, ax=axes[1, 2]).set(title='Produtos de Mercearia - Distribuição')

plt.tight_layout() 
plt.show()
```

## Análise Bivariada para Identificar Padrões

```python
# Relação entre 'Annual Income' e 'Spending Score'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=mall_customers_df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender')
plt.title('Renda Anual vs Pontuação de Gasto')
plt.show()
```

## Análise de Clusters

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Normalizar dados do Mall Customers para segmentação
scaler = StandardScaler()
mall_customers_scaled = scaler.fit_transform(mall_customers_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])

# Aplicar K-Means com 4 clusters (pode ajustar o número de clusters)
kmeans = KMeans(n_clusters=4, random_state=42)
mall_customers_df['Cluster'] = kmeans.fit_predict(mall_customers_scaled)

# Visualizar os clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=mall_customers_df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis')
plt.title('Segmentação de Clientes com K-Means')
plt.show()
```
